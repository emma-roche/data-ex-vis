This lecture covers the transition from static graphical reports to dynamic analytical interfaces, focusing on the theoretical underpinnings and practical implementation using the Python open-source ecosystem.

---

## 1. Theoretical Foundations: The "Why" of Interactivity

* **From Representation to Exploration:**
    * **Static Visualization:** A "fixed narrative" where the author chooses what the reader sees (e.g., a PDF plot).
    * **Interactive Visualization:** A "discovery environment" where the reader becomes the explorer, querying the data in real-time.


* **The "Shneiderman Mantle" of Visual Information Seeking:**
    * *Overview first:* Get a global sense of the physical system (e.g., whole-sky survey).
    * *Zoom and Filter:* Remove noise or focus on specific parameter spaces (e.g., energy ranges).
    * *Details on Demand:* Hover or click to see raw values, uncertainties, or metadata.


* **Cognitive Load and Perceptual Inferences:**
    * Interactivity reduces the need for the brain to "calculate" differences between static frames.

* **Brushing and Linking:** Interacting with a data point in one plot (e.g., a histogram) highlights it in another (e.g., a spatial map), revealing multidimensional correlations.


* **Addressing Transparency in Science:**
    * Static error bars can hide the underlying distribution (Anscombe's Quartet).
    * Interactivity allows peers to "see the data" behind the means, inspecting outliers and raw residuals.



---

## 2. Practical Implementation: The Python Ecosystem

Scientific interactivity in Python generally follows two paths: **Standalone Widgets** (Jupyter-based) and **Web Dashboards** (Deployment-based).

### A. High-Level Declarative Tools (Fast & Intuitive)

* **Altair:** * Based on the "Grammar of Graphics" (Vega-Lite).
* Ideal for statistical data where you want complex interactivity (selections, filtering) with minimal code.


* **HoloViews / PyViz:** * Wraps other libraries like Bokeh or Matplotlib.
* Allows you to "annotate" your data so that the library chooses the best interactive representation automatically.



### B. Interactive Web-Ready Graphics (Standard for Scientists)

* **Plotly:** * The industry standard for interactive 2D and 3D scientific plots.
* Native support for LaTeX rendering in labels and mathematical axes.


* **Bokeh:** * Excellent for large datasets and real-time streaming data (e.g., lab sensor outputs).
* Generates "web-ready" HTML that works in any browser without a Python backend.



### C. Advanced 3D and Spatial Visualization

* **PyVista:** * The "Pythonic" interface for VTK (Visualization Toolkit).
* Essential for physical sciences requiring 3D mesh analysis, finite element models, or volumetric data (CT scans, fluid dynamics).


* **ipyleaflet / Folium:** * Interactive maps for geospatial data, useful in geophysics and environmental monitoring.

---

## 3. Deployment: Sharing Your Results

* **Jupyter Notebooks / Lab:** * The primary environment for "Exploratory Data Analysis" (EDA).
* Use `ipywidgets` to add sliders and buttons directly to your research environment.


* **Streamlit:** * The fastest way to turn a Python script into a sharing-ready web application.
* Requires zero HTML/JS knowledge; perfect for creating a "result viewer" for a published paper.


* **Dash (by Plotly):** * For building more complex, "enterprise-grade" interactive dashboards.

* **Rshiny (R):** * If your team uses R, Shiny is the go-to for interactive web apps, with seamless integration of ggplot2 for visualization.

### Bokeh

is an *interactive visualization library* that targets modern web browsers for presentation. https://bokeh.pydata.org/en/latest/

Let's see an example script:

````python
import numpy as np

## Load bokeh modules

# functions for creating outputs
from bokeh.io import output_file, show, output_notebook

# figure creating library
from bokeh.plotting import figure

# Interactivity tools
from bokeh.models import HoverTool

## create output either in this notebooks output cell or in a separate html file
## either way then result is going to be some html + javascript code 
# output_file("hexbin.html")
output_notebook()

# generate data
n = 500
x = 2 + 2*np.random.standard_normal(n)
y = 2 + 2*np.random.standard_normal(n)

# create an empty figure with some non default settings
p = figure(title="Hexbin for 500 points", match_aspect=True,
           tools="wheel_zoom,reset", background_fill_color='#440154')

# add grid to the figure
p.grid.visible = False

# create a hexbin, which is a 2d pseudocolor histogram on a hexagonal lattice
r, bins = p.hexbin(x, y, size=0.5, hover_color="pink", hover_alpha=0.8)

# add small circles that show each datapoint
p.scatter(x, y, color="white", size=1)

# add interactivity tools: mouse hovering over the data
p.add_tools(HoverTool(
    tooltips=[("count", "@c"), ("(q,r)", "(@q, @r)")],
    mode="mouse", point_policy="follow_mouse", renderers=[r]
))


# create output
show(p)
````
