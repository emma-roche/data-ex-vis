```python
%load_ext sql

# We need this for some reason
# without this we get an error apparently: 
# https://stackoverflow.com/questions/79153112/keyerror-default-when-attempting-to-create-a-table-using-magic-line-sql-in-j
%config SqlMagic.style = '_DEPRECATED_DEFAULT'
```

```python
# This opens a database in the memory
%sql sqlite:////v/courses/2025-202425adatexplorcisvizualizci.public/Lectures/SQL/data/chinook.db
```

```python
# This loads the rows from the tracks table into a ResultSet that can be converted into pandas dataframe and prints it
%sql SELECT * from tracks;  
```

```python
df = %sql SELECT * from tracks;  
df.head()
```
