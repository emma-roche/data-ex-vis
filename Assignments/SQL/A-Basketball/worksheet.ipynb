```python
import sqlalchemy
import pandas as pd
from dotenv import load_dotenv
import os

load_dotenv("/v/courses/2025-202425adatexplorcisvizualizci.public/basketball-mssql.env")
```

```python
# Connection info to the database
server=os.getenv("MSSQL_HOST")
port=os.getenv("MSSQL_PORT")
user=os.getenv("MSSQL_USER") 
password=os.getenv("MSSQL_PASSWORD")
database=os.getenv("MSSQL_DATABASE")
```

```python
# This opens a connection to the MS SQL server using pymssql
# import pymssql
# conn = pymssql.connect(server=server,
#                       port=port,
#                       user=user,
#                       password=password,
#                       database=database)
# This closes it 
# conn.close()
```

```python
# With sqlalchemy
engine = sqlalchemy.create_engine(f"mssql+pymssql://{user}:{password}@{server}/{database}")
connection = engine.connect()
cursor = connection.connection.cursor()
```

### If you need information about the database itself:

```python
# Using the sqlalchemy connection object
res = cursor.execute('select * from INFORMATION_SCHEMA.tables')
print(cursor.fetchall())
```

```python
# A test query using sqlalchemy
res = cursor.execute('select * from player;')
print(cursor.fetchone())
```

```python
# Using pandas
tables = pd.read_sql('select * from INFORMATION_SCHEMA.tables', engine)
columns = pd.read_sql('select * from INFORMATION_SCHEMA.columns', engine)
cols = columns[['TABLE_NAME','COLUMN_NAME', 'DATA_TYPE']]
```
