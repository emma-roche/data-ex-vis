Protein-protein interaction (PPI) networks are scale-free networks. This means that the degree distribution follows a power law. In practice, this can be seen as there are a lot of nodes with very few connection, while there are a few that have relatively high degrees. The latter are usually called *hubs*.
For this task you can use data from the STRING database (https://string-db.org/). STRING is a database of known and predicted protein-protein interactions. The interactions include direct (physical) and indirect (functional) associations; they stem from computational prediction, from knowledge transfer between organisms, and from interactions aggregated from other (primary) databases.
